Here we have Imported the Libraries as per the requirement as we need those to run every thing we need to visualize our data

Here we have imported our data for further visualization

```python
df = pd.read_csv("Salary Data.csv")
df.head()
```

As when we checked if any of our data has a unique/null value we have put down this argument in order to remove the NaN value which was present in the data

```python
df=df.dropna()
df.isnull().sum()
```

```python
df["Years of Experience"].unique()
```

Here we try to figure if our data still has any implict, explicit or unique value in the Salary column

```python
df["Salary"].unique()
```

What is the Highest salary earned by male and females?

Finally after preprocessing of data here we start with the visualization in the following step we have checked how much is salary difference between the male and female in the data

```python
sbrn.barplot(x = "Gender", y = 'Salary', data = df)
matplt.title('Male and Female Salary')
```

How is Years of experience directly proportional to the salary?

Here we have visualized how years of experience have direct effect the the salary of the Males and Females in the data

```python
sbrn.lineplot(x = 'Years of Experience', y = 'Salary', data = df)
matplt.title("Salary With Respect to Years of Experience")
```

What is the amount of people with bachelors, masters and Phd?


Here we have visualized the Education Level of the candidates in the data using a pie chart

```python
df['Education Level'].value_counts().plot(kind='pie')
matplt.title("Education level of every candidiate")
```

What is the highest Salary of Males and Females with respect to their Education Level?

Here we have figured the Highest Salary earned by Males and Females with respect to their education Level and found the salary gap in the between males and females having the same qualification as it also depends on a factor of the total years of experience

```python
sbrn.barplot(x="Salary",y='Education Level', hue="Gender", data=df)
matplt.title("Highest Salaries among Male and Female as per their Education Level")
```

What is the Year Of Experience level compared to Age of the Variable in the data?

Here we check the Years of Experience compared with the Age of the males and females 

```python
sbrn.lmplot(x="Age", y="Years of Experience",data = df,order=1)
matplt.ylabel('Years of Experience')
matplt.xlabel('Age')
matplt.title("Comparing Age with Years of Experience")
```
